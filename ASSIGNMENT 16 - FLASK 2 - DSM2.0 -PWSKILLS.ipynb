### Q1. Explain GET and POST methods.

GET - When you use a web browser to visit a website or click on a link, your browser sends a GET request to the web server, asking for the content of the specified URL.

POST - the data is sent in the request body, which makes it more suitable for sending larger amounts of data or sensitive information. For example, when you submit a form on a website, the form data is typically sent using a POST request.

### Q2. Why is request used in Flask?

'request' object allows you to access and process the data sent by the client (e.g., a web browser) in your Flask application.

The request object allows you to access various data sent by the client, such as form data, URL query parameters, and files uploaded through forms. You can extract this data from the request object and use it in your view functions to process user input.

### Q3. Why is redirect() used in Flask?

redirect() is a function provided by the framework that allows you to redirect a client (usually a web browser) to a different URL. Flask uses routes to map URLs to specific view functions. By using redirect(), you can easily redirect a request to a different endpoint, allowing you to organize your application's logic and flow effectively.

### Q4. What are templates in Flask? Why is the render_template() function used?

Templates enable you to create dynamic HTML pages by combining static content with placeholders for dynamic data. 

The render_template() function is used to render templates in Flask. It takes the name of the template file as its argument and returns the processed HTML content. The function searches for the specified template file in the templates folder.
